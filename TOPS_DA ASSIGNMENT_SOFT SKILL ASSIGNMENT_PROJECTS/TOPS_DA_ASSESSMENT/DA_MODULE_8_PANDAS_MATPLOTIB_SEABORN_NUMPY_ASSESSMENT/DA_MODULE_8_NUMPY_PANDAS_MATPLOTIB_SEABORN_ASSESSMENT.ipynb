#### (7) What are the top 5 acquiring institutions?

```python
Top_5_acquirng_institutions = Banks.groupby('Acquiring Institution').size().sort_values(ascending=False).head()
print("\nTop 5 acquiring institutions:")
print(Top_5_acquirng_institutions)
```

#### (8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?

```python
Banks
```

```python
state_bank_of_texas_acquisitions = Banks[Banks['Acquiring Institution'] == 'State Bank of Texas']
total_acquisitions = len(state_bank_of_texas_acquisitions)
texas_acquisitions = len(state_bank_of_texas_acquisitions[state_bank_of_texas_acquisitions['ST'] == 'TX'])

print("\nNumber of banks acquired by State Bank of Texas:", total_acquisitions)
print("Number of Texas banks acquired by State Bank of Texas:", texas_acquisitions)
```

#### (9) What is the most common city in California for a bank to fail in?

```python
most_common_city_ca = Banks[Banks['ST'] == 'CA']['City'].value_counts().idxmax()
print("The most common city in California for a bank to fail in:", most_common_city_ca)
```
