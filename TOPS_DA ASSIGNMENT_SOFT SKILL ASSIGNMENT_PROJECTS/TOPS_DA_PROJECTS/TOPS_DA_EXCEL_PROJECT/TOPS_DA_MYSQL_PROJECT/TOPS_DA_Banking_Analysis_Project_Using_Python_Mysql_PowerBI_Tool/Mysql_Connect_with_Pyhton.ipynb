```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bins = [0, 100000, 300000, float("inf")]
labels = ['Low', 'Medium', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)
```

```python
df.value_counts('Income Band')
```

```python
df['Income Band'].value_counts().plot(kind='bar', color=['blue', 'purple', 'green'])
```

```python
df.info()
```

```python
categorical_columns = df[["BRId", "GenderId", "IAId", "Income Band", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns
for col in categorical_columns:
    print(f"Value counts for {col}:")   
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(categorical_columns):
    plt.figure(i, figsize=(10, 5))
    sns.countplot(x= predictor, data=df, palette="Set2", hue='GenderId')
    plt.title(f"Count of {predictor}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
for i, predictor in enumerate(categorical_columns):
    plt.figure(i, figsize=(10, 5))
    sns.countplot(x= predictor, data=df, palette="Set2", hue='Nationality')
    plt.title(f"Count of {predictor}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in categorical_columns:
    if col == "ocupation":
        continue
    plt.figure(figsize=(10, 5)) 
    sns.histplot(df[col], bins=20, color='blue', alpha=0.7)
    plt.title(f"Histogram of {col}")
```
